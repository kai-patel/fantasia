```python
import pandas as pd
import json
```

```python
with open("data.json", 'r') as f:
    data = json.load(f)
    
print(data.keys())
```

```python
print(data['elements'][0])
print(len(data['elements'][0]))
```

```python
from pandas.io.json import json_normalize
df = json_normalize(data['elements'])
df = df.set_index('web_name')
```

```python
dropcols = ["id", "bonus", "element_type", "saves", "first_name", "photo", "news_added", "dreamteam_count", "transfers_out_event", "special", "squad_number", "status", "team", "team_code", "second_name", "code", "news", 'influence_rank', 'influence_rank_type', 'creativity_rank', 'creativity_rank_type', 'threat_rank', 'threat_rank_type', 'ict_index_rank', 'ict_index_rank_type', 'corners_and_indirect_freekicks_text', 'direct_freekicks_text', 'penalties_text']
clean = df.drop(dropcols, axis=1)

clean = clean.dropna(axis=1, how='any')
```

```python
%matplotlib inline

import sys
import matplotlib.pyplot as plt

fig, axes = plt.subplots(figsize=(8, 120), dpi= 100)
clean.tail(2).plot(kind='line', y=None, sharex=False, subplots=True, ax=axes, legend=False, stacked=True)

#print(clean.head(1).shape)

plt.subplots_adjust(left = 0.125, right = 0.9, bottom = 0.1, top = 0.9, wspace = 0.2, hspace = 0.7)
#plt.tight_layout()
```
